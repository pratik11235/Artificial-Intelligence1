```python
import pandas as pd

# Read file contents and store them in pandas dataframe with required headers
headers = ['baseball_game_on_TV', 'George_watches_TV', 'out_of_cat_food', 'George_feeds_cat']
data = []

with open('training_data.txt', 'r') as f:
    line = f.readline().rstrip('\n')
    data.append(line.split('     ')[1:])
    while line:
        line = f.readline().rstrip('\n')
        if line != "":
            data.append(line.split('     ')[1:])

df = pd.DataFrame(data, columns = headers)
df
```

```python
# Calculating prior probability of Baseball_game_on_TV
count_true = 0
tot_count = 0

for i in df['baseball_game_on_TV'] == '1':
    if i == True:
        count_true += 1
    tot_count += 1

print(count_true, tot_count)
P1 = count_true/tot_count
print("P(baseball_game_on_TV) =", P1)
```
